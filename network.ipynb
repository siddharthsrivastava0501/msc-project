```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rc('animation', html='jshtml')

Vr = -65
I = 20
R = 1
tau = 5
V_thresh = -50
V_spike = 30

T = 50
dt = .1
t = np.arange(0, T + dt, dt)
z = np.zeros(len(t))
z[0] = Vr

# z = [v(t)]
# f(z) = [dv/dt]
def f(t, z):
    dv = (Vr - z + R*I)/tau
    return np.array([dv])

for i in range(len(t) - 1):
    if z[i] >= V_thresh: 
        z[i] = V_spike
        z[i + 1] = Vr
    else:
        z[i + 1] = z[i] + dt * f(t[i], z[i])

plt.figure(figsize=(5.31,3))
plt.plot(z)
plt.xticks([0,100,200,300,400,500], [0,10,20,30,40,50])
plt.ylim(-70, 45)
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Potential (mV)')
plt.title(r'LIF neuron spiking dynamics')
plt.grid(linestyle = '-.', linewidth=0.5)
plt.savefig('LIF_neuron.pdf', dpi=600, bbox_inches='tight')
```
